```python
import numpy as np
```

```python
from scipy. integrate import odeint 
```

```python
import matplotlib.pyplot as plt
```

```python
import math
```

```python
a = 0.01
b = 0.02
```

```python
N = 5555
I0 = 75
R0 = 4
```

```python
S0 = N - I0 - R0
```

```python
def syst(t, x):
    t1, t2, t3 = t
    return [0, -b*t2, b*t2 ]
```

```python
def syst2(t, x):
    t1, t2, t3 = t
    return [-a*t1, a*t1-b*t2, b*t2 ]
```

```python
x0 = 0
```

```python
y0 = [S0,I0,R0]
```

```python
x = np.arange( 0, 200, 0.01)
t1 = odeint(syst, y0, x)
t1s = t1[:,0]
t1i = t1[:,1]
t1r = t1[:,2]
```
