```python
import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt
import math
```

```python
t0 = 0
x0 = 23

N = 3010
```

```python

a1 = 0.566
a2 = 0.000066
```

```python
t = np.arange(t0, 20, 0.1)
```

```python
def syst(dx, t):
    x = dx
    return (a1 + x*a2)*(N-x)
```

```python
y = odeint (syst, x0, t)
```

```python
fig1 = plt.figure(facecolor = 'white')
plt.plot(t, y, linewidth = 2, label = "Решение")
plt.xlabel("t")
plt.ylabel("численность")
plt.grid()
plt.legend()
plt.show()
fig1.savefig('01.png', dpi = 600)
```

```python
a1 = 0.000044
a2 = 0.244
```

```python
t = np.arange(t0, 0.5, 0.01)
```

```python
y = odeint (syst, x0, t)
dy = (a1 + y*a2)*(N-y)
```

```python
fig2 = plt.figure(facecolor = 'white')
plt.plot(t, y, linewidth = 2, label = "Решение")
plt.plot(t, dy, linewidth = 2, label = "Производная")
plt.xlabel("t")
plt.ylabel("численность")
plt.grid(True)
plt.legend()
plt.show()
fig2.savefig('02.png', dpi = 600)
```

```python
def a1(t):
    a1 = 0.44*t
    return a1
```

```python
def a2(t):
    a2 = 0.34*math.cos(t)
    return a2
```

```python
t = np.arange(t0, 0.1, 0.01)
```

```python
def syst2(dx, t):
    x = dx
    return (a1(t) + x*a2(t))*(N-x)
```

```python
y = odeint(syst2, x0, t)
```

```python
fig3 = plt.figure(facecolor = 'white')
plt.plot(t, y, linewidth = 2, label = "Решение")
plt.xlabel("t")
plt.ylabel("численность")
plt.grid(True)
plt.legend()
plt.show()
fig3.savefig('03.png', dpi = 600)
```
